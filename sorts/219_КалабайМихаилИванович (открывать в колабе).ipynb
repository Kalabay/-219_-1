```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("result.csv")
df
```

Графики интерактивные, можно приближать!!! Всё в наносекундах!!!

```python
import plotly.express as px

fig = px.line(df[df['type_array'] == '50-300 step=50 x = 0...5'], x="n", y="time", color='type_sort', title='50-300 step=50 x = 0...5')
fig.show()
```

Можно заметить, что все пузырьки ведут себя плохо, оптимизации делу не помогают, что логично, так как массив случайный. Бинарный insert хуже, потому что всё равно квадрат, но и ещё плюс самописный бинпоиск. Понятно, что все nlogn выигрывают за исключением heap, потому что там большая константа из-за самописной реализации и самого алгоритма сортировки. Ну и абсолютная доминация сортировки подсчётом, потому что O(n + maxn), где maxn всего лишь 5. Insert лучшая из O(n^2)

```python
import plotly.express as px

fig = px.line(df[df['type_array'] == '50-300 step=50 x = 0...4000'], x="n", y="time", color='type_sort', title='50-300 step=50 x = 0...4000')
fig.show()
```

Отличие от прошлого вывода понятно, countSort перестал быть лучшим, потому что maxn = 4000

```python
import plotly.express as px

fig = px.line(df[df['type_array'] == '50-300 step=50 almost (+- 90% sort)'], x="n", y="time", 
              color='type_sort', title='50-300 step=50 almost (+- 90% sort)')
fig.show()
```

Оптимизации bubble работают! (из-за того, что почти отсорт, быстро выходит) И понятно, что 1+2 лучше. По этой же причине вставки очень хороши. А вот быстрая страдает из-за первого опорного (то есть зачастую min)

```python
import plotly.express as px

fig = px.line(df[df['type_array'] == '50-300 step=50 reverse'], x="n", y="time", 
              color='type_sort', title='50-300 step=50 reverse')
fig.show()
```

Selection работает лучше других n^2, так как min из-за reverse находится быстро. Остальным n^2 очень тяжело, очевидно, что оптимизации bubble не работают тут

```python
import plotly.express as px

fig = px.line(df[df['type_array'] == '100-4000 step=100 x = 0...5'], x="n", y="time", 
              color='type_sort', title='100-4000 step=100 x = 0...5')
fig.show()
```

Типичная ситуация. На 5 элементах count лучшая, heap проигрывает остальным nlogn из-за константы, однако лучше n^2, пузыри худшие и т.д.

У Quick те же проблемы, что и на 300 (и у bubble тоже). Важные наблюдения: Shell с Ciura постоянно доминирует над Shell с Shell. MergeSort - самая стабильная

```python
type1 = ["50-300 step=50 x = 0...5", "50-300 step=50 x = 0...4000", "50-300 step=50 almost (+- 90% sort)", "50-300 step=50 reverse"]
type2 = ["100-4000 step=100 x = 0...5", "100-4000 step=100 x = 0...4000", "100-4000 step=100 almost (+- 90% sort)", "100-4000 step=100 reverse"]
```

```python
mask_type2 = (df['type_array'] == type2[0]) | (df['type_array'] == type2[1]) | (df['type_array'] == type2[2]) | (df['type_array'] == type2[3])
mask_type1 = (df['type_array'] == type1[0]) | (df['type_array'] == type1[1]) | (df['type_array'] == type1[2]) | (df['type_array'] == type1[3])
```

```python
import plotly.express as px

fig = px.line(df[(df['type_sort'] == 'selectionSort') & mask_type1], x="n", y="time", 
              color='type_array', title='50-300 selectionSort')
fig.show()
```

Всё +- стабильно

```python
import plotly.express as px

fig = px.line(df[(df['type_sort'] == 'selectionSort') & mask_type2], x="n", y="time", 
              color='type_array', title='100-4000 selectionSort')
fig.show()
```
